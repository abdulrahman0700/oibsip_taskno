# Exploratory Data Analysis


```python
fig , ax = plt.subplots(2,1 ,figsize=(7,6))
sns.lineplot(data,x='Date',y=' Estimated Employed',hue="Area",ax=ax[0])
ax[0].set_xlabel("Time")

sns.lineplot(data,x='Date',y=' Estimated Unemployment Rate (%)',hue='Area',ax=ax[1])
ax[1].set_xlabel("Time")

plt.show()
```

```python
sns.lineplot(data,x='Date',y=' Estimated Labour Participation Rate (%)',hue="Area")
plt.xlabel('Time')
plt.show()
```

```python
fig , axes = plt.subplots(1,2 , figsize=(15,8))
sns.barplot(data,x='Area',y=' Estimated Unemployment Rate (%)',ax=axes[0])
sns.barplot(data,x='Area',y=' Estimated Labour Participation Rate (%)',ax=axes[1])
sns.set_theme(style='dark')
```

```python
plt.subplots(1,figsize=(15,8))
sns.barplot(data,x=' Estimated Labour Participation Rate (%)',y='Region',hue='Area')
plt.show()
```

```python
fig , ax = plt.subplots(2,1,figsize=(14,9))
sns.boxplot(data,x=' Estimated Labour Participation Rate (%)',y='Area',ax=ax[0])
sns.boxplot(data,x=' Estimated Unemployment Rate (%)',y='Area',ax=ax[1])
sns.set_theme(style='darkgrid')
```

```python
plt.subplots(1,figsize=(16,13))
sns.boxplot(data,x=' Estimated Labour Participation Rate (%)',y='Region',hue='Area')
plt.show()
```

```python
plt.subplots(1,figsize=(16,11))
sns.histplot(data,y='Region',stat='probability',hue='Area',kde=True ,multiple="stack")
plt.show()
```
